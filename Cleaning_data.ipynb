## Cleaning Chicago Marathon Data

Data from this will be used in Tableau

```python
#Pandas library used for cleaning
import pandas as pd
```

```python
#Importing csv with pace of all runners
df = pd.read_csv('pace_testset.csv')
```

```python
df.head()
```

```python
#Dropping unecessary column
df = df.drop(columns =['Unnamed: 0'])
```

```python
#Importing full unencoded chicago marathon results
df2 = pd.read_csv('full_uneconded_results.csv')
```

```python
df2.head()
```

```python
df2 = df2[['Start Time', 'Finish Time of Day', 'Finish Time.1']]
```

```python
df2.head()
```

```python
result = pd.concat([df, df2], axis=1)
```

```python
result.head()
```

```python
#Sorting values by Overall Place
df = result.sort_values(by=['Place Overall'])
```

```python
#Testing if it works
df.tail(50)
```

```python
#Testing if it is sorted by Place Overall
df.head(50)
```

```python
#Saving into a final csv
df.to_csv("final.csv")
```
